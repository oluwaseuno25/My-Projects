```python
#Question1

# What is the age distribution of  people over the decade?

#question 2(static visualization)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)

population = pd.read_csv('https://raw.githubusercontent.com/suthapalliuday/INFO5502-Summer2022/main/datasets/lab_assignment_05_census.csv')
print(population.head())
plt.figure(figsize = (12,12))
plt.title("Age Distribution of  people over the 2 years")
plt.grid(color = 'green', linestyle = '--', linewidth = 1.0)
sns.scatterplot(data = population, x = 'Age', y = 'People', hue = 'Sex')

```
