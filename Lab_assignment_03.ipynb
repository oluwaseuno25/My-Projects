```python
import pandas as pd
import random
import namegenerator

from faker import Faker
from random import randrange
from datetime import datetime

no_of_movies = 1000

fake = Faker('de_DE')

movies = []

for movies_id in range(no_of_movies):
    # Create movie name
    title = (namegenerator.gen())

    # Create release date 
    d1 = datetime.strptime(f'1/1/2000', '%m/%d/%Y')
    d2 = datetime.strptime(f'8/7/2022', '%m/%d/%Y')
    release_date = fake.date_between(d1, d2)

    #create director's name
    name = fake.name()

    # Create genre
    genre = random.choice(["Drama", "Fantasy", "Action.", "Comedy","Horror", "Romance", "Thriller.", "Mystery" ])

    # Create movie ratings 
    ratings = fake.pyfloat(right_digits=2, positive=True, min_value=1, max_value=10)

    #Create language
    language = random.choice(["English", "French", "Hindi.", " Japanese","Spanish", "Italian", "Russia", "Mandarin", "German" ])

    #create viewers vote
    vote = fake.ean(length=8)
    
    #create amount spent
    budget = fake.pyfloat(right_digits=00, positive=True, min_value=2000, max_value=250000)

    movies.append([title, release_date, name, genre, ratings, language, vote, budget])

movies_df = pd.DataFrame(movies, columns=['Movie Title','Released_date','Directors Name', 'Genre','Movie ratings', 'Language','Viewers vote', 'Budget ($)']) 
                
pd.pandas.set_option('display.max_columns', None)
print(movies_df)
```
