```python
#To generate 12 random numbers between 1 and 100 and sort the list in ascending and descending number,Average calculated for the group
import random
list=[]
for i in range(12):
    num=random.randint(1,100)
    list.append(num)
    print(list[i], end=" ")
list.sort()
#ascending order
print("List in Ascending Order:",list)
list.sort(reverse=True)
print('List in Descending Order: ', list)

# Calculate average values for ascending order values
def Average(l): 
    avg = sum(l) / len(l) 
    return avg
my_list1 = [7, 8, 11, 12, 16, 17, 19, 34, 44, 49, 73, 82] 
average = Average(my_list1) 
print("Average of my_list1 is", average)
# Calculate average values for descending order values

from statistics import mean

def Average(l): 
    avg = mean(l) 
    return avg
  
my_list2 = [91, 85, 82, 80, 72, 36, 20, 18, 18, 14, 3, 2] 
average = Average(my_list2) 
  
print ("Average of my_list_2 is", average)



```
